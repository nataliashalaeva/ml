```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('AirQualityUCI.csv', sep=';')
```

```python
print(df.head(10))
```

```python
df['NOx(GT)'] = df['NOx(GT)'].str.replace(',', '.', regex=True).astype(float)
```

```python
df['NOx(GT)'] = df['NOx(GT)'].replace(-200, pd.NA)
```

```python
df['NOx(GT)'] = pd.to_numeric(df['NOx(GT)'], errors='coerce')
```

1. Найдите количество наблюдений для каждого диапазона значений целевой переменной
"CO(GT)" (уровень концентрации оксида углерода). Разбейте значения CO(GT) на несколько
интервалов (например, с помощью pd.cut или pd.qcut) и примените метод plot к полученным
данным, чтобы построить столбчатую диаграмму, показывающую распределение концентраций
CO.

```python

bins = [0, 1, 2, 3, 5, 7, 10, 15, 20, 50]
labels = ['0-1', '1-2', '2-3', '3-5', '5-7', '7-10', '10-15', '15-20', '20-50']
df['CO_bins'] = pd.cut(df['CO(GT)'], bins=bins, labels=labels, right=False)
co_distribution = df['CO_bins'].value_counts().sort_index()
co_distribution.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Распределение CO(GT)')
plt.xlabel('Диапазоны CO')
plt.ylabel('Количество наблюдений')
plt.show()
```

2. Постройте такую же столбчатую диаграмму, как в задаче 1, но используйте логарифмический
масштаб для оси Y (количества наблюдений). Это поможет лучше визуализировать распределение,
если есть очень большие различия в частоте встречаемости разных диапазонов концентраций CO.

```python
co_distribution.plot(kind='bar', color='lightgreen', edgecolor='black', log=True)
plt.title('CO(GT) с логарифмической шкалой')
plt.xlabel('Диапазоны CO')
plt.ylabel('Логарифмическая частота')
plt.show()
```

3. Постройте две гистограммы по значениям признака "T" (температура) – одну для наблюдений,
где уровень концентрации оксида углерода (CO(GT)) выше среднего, и другую – где он ниже
среднего.

```python
mean_co = df['CO(GT)'].mean()

high_co = df[df['CO(GT)'] > mean_co]['T']
low_co = df[df['CO(GT)'] <= mean_co]['T']

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(low_co.dropna(), bins=20, color='pink', alpha=0.7, edgecolor='black')
plt.title('Температура при низком CO(GT)')
plt.xlabel('Температура (°C)')
plt.ylabel('Частота')

plt.subplot(1, 2, 2)
plt.hist(high_co.dropna(), bins=20, color='grey', alpha=0.7, edgecolor='black')
plt.title('Температура при высоком CO(GT)')
plt.xlabel('Температура (°C)')
plt.ylabel('Частота')

plt.tight_layout()
plt.show()
```

4. В задаче 3 подберите значение аргумента density=True в функции hist так, чтобы по вертикали
графика была расположена не частота, а плотность распределения. Число бинов должно равняться
20 для обеих гистограмм, а коэффициент alpha равным 0.5, чтобы гистограммы были
полупрозрачными и не загораживали друг друга.

```python
mean_co = df['CO(GT)'].mean()
df[df['CO(GT)'] > mean_co]['T'].plot.hist(bins=20, alpha=0.5, density=True, color='red', label='CO выше среднего')
df[df['CO(GT)'] <= mean_co]['T'].plot.hist(bins=20, alpha=0.5, density=True, color='blue', label='CO ниже среднего')
plt.xlabel('Температура (°C)')
plt.ylabel('Плотность распределения')
plt.title('Гистограммы температуры')
plt.show()
```

6. Аналогично изобразите гистограмму распределения уровня концентрации бензола (C6H6(GT))
относительно времени суток (можно выделить, например, утренние, дневные, вечерние и ночные
часы на основе признака "Время").


```python
df['Time'] = df['Time'].str.split('.').str[0].astype(int)
```

```python
def time_of_day(hour):
    if 6 <= hour < 12:
        return 'Утро'
    elif 12 <= hour < 18:
        return 'День'
    elif 18 <= hour < 22:
        return 'Вечер'
    else:
        return 'Ночь'

df['TimeOfDay'] = df['Time'].apply(time_of_day)
df['C6H6(GT)'] = df['C6H6(GT)'].fillna(df['C6H6(GT)'].mean())
plt.figure(figsize=(12, 6))

plt.hist(df[df['TimeOfDay'] == 'Утро']['C6H6(GT)'], bins=20, alpha=0.5, label='Утро', color='orange', density=True)
plt.hist(df[df['TimeOfDay'] == 'День']['C6H6(GT)'], bins=20, alpha=0.5, label='День', color='green', density=True)
plt.hist(df[df['TimeOfDay'] == 'Вечер']['C6H6(GT)'], bins=20, alpha=0.5, label='Вечер', color='blue', density=True)
plt.hist(df[df['TimeOfDay'] == 'Ночь']['C6H6(GT)'], bins=20, alpha=0.5, label='Ночь', color='purple', density=True)

plt.title('Распределение уровня концентрации бензола (C6H6(GT)) по времени суток')
plt.xlabel('Концентрация бензола (C6H6(GT))')
plt.ylabel('Плотность распределения')
plt.legend()
plt.show()
```

8. Постройте несколько гистограмм или диаграмм рассеяния, показывающих зависимость
концентраций разных загрязняющих веществ (например, CO(GT), C6H6(GT), NOx(GT)) от
температуры (T). Это поможет оценить, как температура влияет на различные типы загрязнения.

```python
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.scatter(df['T'], df['CO(GT)'], color='blue', alpha=0.5)
plt.title('CO(GT) vs Температура')
plt.xlabel('Температура (°C)')
plt.ylabel('CO(GT)')

plt.subplot(1, 3, 2)
plt.scatter(df['T'], df['C6H6(GT)'], color='green', alpha=0.5)
plt.title('C6H6(GT) vs Температура')
plt.xlabel('Температура (°C)')
plt.ylabel('C6H6(GT)')

plt.subplot(1, 3, 3)
plt.scatter(df['T'], df['NOx(GT)'], color='red', alpha=0.5)
plt.title('NOx(GT) vs Температура')
plt.xlabel('Температура (°C)')
plt.ylabel('NOx(GT)')

plt.tight_layout()
plt.show()
```

9. Постройте график с областями (area plot), показывающий изменение во времени нескольких
показателей, например, концентрации разных веществ. Это позволит наглядно увидеть вклад
каждого вещества в общий уровень загрязнения.

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

plt.figure(figsize=(12, 6))
plt.fill_between(df['Date'], df['CO(GT)'], color='green', alpha=0.5, label='CO(GT)')
plt.fill_between(df['Date'], df['C6H6(GT)'], color='red', alpha=0.5, label='C6H6(GT)')
plt.fill_between(df['Date'], df['NOx(GT)'], color='blue', alpha=0.5, label='NOx(GT)')

plt.title('Изменение концентрации загрязняющих веществ во времени')
plt.xlabel('Дата')
plt.ylabel('Концентрация вещества')

plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
